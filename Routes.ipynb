```python
#andhra routes
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException, TimeoutException
from selenium.webdriver.common.action_chains import ActionChains

# Initialize the WebDriver
driver = webdriver.Chrome()
driver.get('https://www.redbus.in/online-booking/apsrtc/?utm_source=rtchometile')
time.sleep(3)
driver.maximize_window()

# Wait
wait = WebDriverWait(driver, 20)

def Andhra(path):
    routes = []
    route_links = []
    page_number = 1

    while True:
        print(f"Scraping page {page_number}...")
        paths = driver.find_elements(By.XPATH, path)

        for route in paths:
            routes.append(route.text)
            route_links.append(route.get_attribute("href"))

        # Try to click the next page number
        try:
            pagination = wait.until(EC.presence_of_element_located((By.XPATH, '//*[@class="DC_117_paginationTable"]')))


            next_page_number = pagination.find_element(By.XPATH, f'.//div[text()="{page_number + 1}"]')
            if next_page_number:
                print(f"Next page number found: {next_page_number.text}")

                # Scroll into view before clicking
                driver.execute_script("arguments[0].scrollIntoView(true);", next_page_number)
                time.sleep(1)  

                # Using ActionChains to click the next page number
                actions = ActionChains(driver)
                actions.move_to_element(next_page_number).click().perform()
                time.sleep(2)  

                page_number += 1 
            else:
                print("No next page number found.")
                break
        except NoSuchElementException:
            print("No more pages to paginate.")
            break
        except TimeoutException:
            print("Timeout while waiting for pagination.")
            break
        except Exception as e:
            print(f"An error occurred: {e}")
            break

    return routes, route_links

# Call the function
try:
    routes, route_links = Andhra("//a[@class='route']")
finally:
    driver.quit()


for route_name, route_link in zip(routes, route_links):
    print(f"Route Name: {route_name}, Route Link: {route_link}")
```

```python
df_aa = pd.DataFrame({"Route Name":routes , "Route Link": route_links})
df_aa
```

```python
df_aa.to_csv('aa',index=False)
```
